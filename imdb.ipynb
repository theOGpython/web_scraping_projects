```python
#import libraries
from bs4 import BeautifulSoup as bs
import requests
import re
import pandas as pd
```

```python
#Download IMDB top 250 movies data
url = 'http://www.imdb.com/chart/top'
```

```python
response = requests.get(url)
soup = bs(response.text, 'lxml')
```

```python
movies = soup.select('td.titleColumn')
crew = [a.attrs.get('title') for a in soup.select('td.titleColumn a')]
ratings = [b.attrs.get('data-value') for b in soup.select('td.posterColumn span[name=ir]')]
```

```python
cols = ["movie_title", "year", "place", "star_cast", "rating", "director"]
imdb_df = pd.DataFrame(columns=cols)
```

```python
# Store each item into dictionary (data), then put those into a dataframe (imdb_df)
for index in range(0, len(movies)):
       
# Seperate movie into: 'place', 'title', 'year'
    movie_string = movies[index].get_text()
    movie = (' '.join(movie_string.split()).replace('.', ''))
    movie_title = movie[len(str(index))+1:-7]
    year = re.search('\((.*?)\)', movie_string).group(1)
    place = movie[:len(str(index))-(len(movie))]
    director =  crew[index].split(",")[0][:-7]
    data = {"movie_title": movie_title,
            "year": year,
            "place": place,
            "star_cast": ', '.join(crew[index].split(",")[1:]),
            "rating": ratings[index],
            "director": director}
    print(f"Appending Number {index}")
    imdb_df = imdb_df.append(data, ignore_index=True)
```

```python
imdb_df.info()
```

```python
#Change data types
imdb_df["rating"] = pd.to_numeric(imdb_df["rating"])
imdb_df.year.astype('int32')
imdb_df["year"] = pd.to_datetime(imdb_df["year"], format='%Y')
```
